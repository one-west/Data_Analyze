```python
# 3. 본문의 붓꽃 데이터 셋에서 다음 문제를 해결해 보세요.
## 3-1. 꽃 받침 너비의 값을 내림차순으로 정렬해 보세요.
iris = sns.load_dataset("iris")
iris.sort_values("sepal_width", ascending=False)
```

```python
## 3-2. (3-1)의 결과 상위 10개의 데이터를 막대 그래프로 나타내 보세요.
import matplotlib.pyplot as plt

newDf = iris.sort_values(by="sepal_width", ascending=False)
top10 = newDf.head(10)
top10.plot(kind="bar", rot=0)

plt.title("상위 10")
plt.xlabel("Data")
plt.ylabel("Length")
plt.show()
```

```python
## 3-3. 꽃잎의 길이와 꽃잎의 너비의 산점도 그래프를 그려 보세요.
newDf.plot(kind="scatter", x="petal_length", y="petal_width")
```

```python
# 4. 본문의 타이타닉 데이터 셋에서 다음 문제를 해결해 보세요.
## 4-1. 성별에 따른 생존율을 비교해 보세요.
import pandas as pd

titanic = pd.read_csv(resourcePath + "titanic.csv")
titanic[["Sex", "Survived"]].groupby("Sex").mean()
```

```python
## 4-2. 함께 승선한 부모와 자식 수(Parch)에 따른 생존율 비교해 보세요.
titanic[["Parch", "Survived"]].groupby("Parch", as_index=False).mean().sort_values(
    by="Survived", ascending=False
)
```

```python
## 4-3. 객실 등급(Pclass), 성별(Sex), 컬럼 별로 titanic 내 전체 데이터의 평균값을 출력해 보세요.
viewTable = [
    "Pclass",
    "Sex",
    "PassengerId",
    "Survived",
    "Age",
    "SibSp",
    "Parch",
    "Fare",
]

titanic[viewTable].groupby(["Pclass", "Sex"]).mean()
```

```python
## 4-4. 항구별 생존자를 데이터 시각화로 확인해 보세요.

import seaborn as sns

sns.countplot(data=titanic, x="Embarked", hue="Survived")
plt.show()
```
