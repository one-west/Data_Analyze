```python
# 2. '서울특별시 코로나 19 백신 예방접종 현황'에 대한 정보를 가지고 있는 데이터 파일(vaccine.csv)을 이용하여 다음 문제를 차례대로 수행하세요.

# 실행결과 1
file = pd.read_csv("../resources/5장/vaccine.csv", encoding="cp949")
file.tail()
```
