```python
from pandas import *
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

happy_life = read_excel("../resources/9장/대한민국행복지도_삶의만족도.xlsx")
happy_health = read_excel("../resources/9장/대한민국행복지도_건강.xlsx")
happy_safe = read_excel("../resources/9장/대한민국행복지도_안전.xlsx")
happy_environ = read_excel("../resources/9장/대한민국행복지도_환경.xlsx")
happy_econo = read_excel("../resources/9장/대한민국행복지도_경제.xlsx")
happy_edu = read_excel("../resources/9장/대한민국행복지도_교육.xlsx")
happy_relation = read_excel("../resources/9장/대한민국행복지도_관계및사회참여.xlsx")
happy_leisure = read_excel("../resources/9장/대한민국행복지도_여가.xlsx")
```

```python
city = list(happy_life["시도"].unique())
happy_merge = DataFrame({"시도": city})
happy_merge
```

```python
life = happy_life["삶의 만족도"].groupby(by=happy_life["시도"]).mean()
happy_merge = merge(happy_merge, life, on="시도")
happy_merge.head()
```

```python
health = happy_health['평균'].groupby(by=happy_health['시도']).mean() # 시도별 건강 평균
happy_merge = merge(happy_merge, health.rename('건강'), on='시도') # 시도별 건강 평균을 happy_merge에 추가
happy_merge.head()
```

```python
safe = happy_safe['평균'].groupby(by=happy_safe['시도']).mean()
happy_merge = merge(happy_merge, safe.rename('안전'), on='시도')
happy_merge.head()
```

```python
environ = happy_environ['평균'].groupby(by=happy_environ['시도']).mean()
happy_merge = merge(happy_merge, environ.rename('환경'), on='시도')
happy_merge.head()
```

```python
econo = happy_econo['평균'].groupby(by=happy_econo['시도']).mean()
happy_merge = merge(happy_merge, econo.rename('경제'), on='시도')
happy_merge.head()
```

```python
edu = happy_edu['평균'].groupby(by=happy_edu['시도']).mean()
happy_merge = merge(happy_merge, edu.rename('교육'), on='시도')
happy_merge.head()
```

```python
relation = happy_relation['평균'].groupby(by=happy_relation['시도']).mean()
happy_merge = merge(happy_merge, relation.rename('관계및사회참여'), on='시도')
happy_merge.head()
```

```python
leisure = happy_leisure['평균'].groupby(by=happy_leisure['시도']).mean()
happy_merge = merge(happy_merge, leisure.rename('여가'), on='시도')
happy_merge
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import *

# import koreanize_matplotlib # 한글폰트 적용

plt.figure(figsize=(15, 6))
rcParams["font.family"] = "gulim"

items = list(happy_merge.columns[1:9])
for a in items:
    chartdata = happy_merge[a]  # 행복지수 항목별 값
    plt.plot(chartdata, marker="o", label=a)  # 항목별 값 그래프로 표시

plt.xlabel("시도")
plt.ylabel("수치")
plt.xticks(range(17), happy_merge["시도"], fontsize=8)  # x축 레이블
plt.title("대한민국 행복지수")
plt.legend()
plt.grid()
plt.show()
```

```python
from numpy import *

happy_merge.plot(
    kind="bar",
    xlabel="city",
    ylabel="value",
    figsize=(15, 20),
    grid=True,
    subplots=True,  # 각 항목별로 그래프 따로 그림
)

plt.tight_layout(pad=8, h_pad=2)

plt.xlabel("시도")
plt.ylabel("수치")
plt.xticks(range(17), happy_merge["시도"], fontsize=8)  # x축 레이블
plt.title("대한민국 행복지수")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 15))

corr_mat = happy_merge.corr(numeric_only=True)
upp_mat = triu(corr_mat)  # tril() 아래 삼각형

sns.heatmap(data=corr_mat, mask=upp_mat, annot=True, cmap="RdYlGn")
plt.show()
```
