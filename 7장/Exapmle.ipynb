```python
data = pd.read_excel('../resources/7ìž¥/fine_dust.xlsx' , index_col='area')
data2018 = data[2018]
data2018
```

```python
plt.figure(figsize=(15, 4))
plt.plot(data2018, color='b', marker='o')
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 Fine Dust Line Graph')
plt.show()
```

```python
plt.figure(figsize=(15, 4))
for year in range(2016, 2020):
    chartData = data[year]
    plt.plot(chartData, marker="s", label=year)

plt.legend()
plt.xlabel("area")
plt.ylabel("micrometer")
plt.title("2016 ~ 2019 Fine Dust Line Graph")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(15, 4))
plt.bar(x=data2018.index, height=data2018, color='g')
plt.ylim(36, 50)
plt.xlabel('area')
plt.ylabel('micrometer')
plt.title('2018 Fine Dust Line Graph')
plt.grid()
plt.show()
```

```python
import numpy as np

index = np.arange(4)
plt.figure(figsize=(15, 4))
data4 = data.loc["Gyeonggi":"Daegu", 2016:2019]

for year in range(2016, 2020):
    chartData = data4[year]
    plt.bar(x=index, height=chartData, width=0.2, label=year)
    index = index + 0.2
    # 2016 : xì¶• -> 0, 1, 2, 3
    # 2017 : xì¶• -> 0.2, 1.2, 2.2, 3.2
plt.legend()
plt.xlabel("area")
plt.ylabel("micrometer")
plt.title("2016 ~ 2019 Fine Dust Bar Graph")
plt.xticks(index-0.5, ["Gyeonggi", "Incheon", "Busan", "Daegu"])
plt.grid()
plt.show()
```
